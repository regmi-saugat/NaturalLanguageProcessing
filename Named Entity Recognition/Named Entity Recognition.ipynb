# Named Entity Recognition
                                
                                    - Using `spaCy` library

```python
#@ Loading the required libraries
import spacy
import spacy.displacy as displacy
```

### Using Small Sized Model

```python
nlp = spacy.load("en_core_web_sm")

doc1 = nlp("""Nepal lies between China and India in South Asia. The country is slightly larger than the state of 
           Arkansas. Nepal has the greatest altitude change of any location on Earth. The lowlands are at sea level 
           and the mountains of the Himalaya are the tallest in the world. Mount Everest rises to 29,035 feet 
           (8,850 meters) and is the world’s highest peak.The Himalaya formed 10–15 million years ago when India 
           collided with the continent of Asia and pushed the land into high mountains. Eight of the world’s ten 
           highest mountain peaks are in Nepal.""")
for token in doc1:
    print(token.text, token.pos_, token.dep_)
```

```python
#@ Visualizing the entities recognized by the document
displacy.render(doc1, style='ent', jupyter=True)
```

```python
#@ Visualizing the given document 
displacy.render(doc1, style='dep', jupyter=True)
```
